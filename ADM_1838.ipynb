```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([2,3,9,12])
ypoints=np.array([5,7,10,15])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,41,3,105,7])
plt.plot(ypoints,marker='o',ms=20,mec='royalblue',mfc='yellow')
plt.plot(ypoints,ls=':')
plt.plot(ypoints,color='r')
plt.plot(ypoints,linewidth='20.5')

plt.plot(ypoints, "p:y")
plt.show()
```
